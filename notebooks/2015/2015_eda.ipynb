```python
""" get a genes """

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 8, 10
fig, axs = plt.subplots(4, 2)

def smooth_rows(df, n_points=100, power=3, standardize=True):
    """A function to smooth the rows of a dataframe

    args:
        : df (dataframe): m observations by n timepoints
        : n_points (int): number of points to interpolate over
        : power (int): power of the spline
    """

    S = np.zeros((df.shape[0], n_points))

    for i, (idx, row) in enumerate(df.iterrows()):
        
        y = row.values  
        x = list(range(len(y)))
        x_hat = np.linspace(0, len(y), n_points) 

        spl = make_interp_spline(x, y, k=power)  
        smooth_row = spl(x_hat)
        if standardize:
            smooth_row =  (smooth_row - smooth_row.mean())/(smooth_row.std())
        S[i, :] = smooth_row
    
    S = pd.DataFrame(S, index=df.index, columns=x_hat)
    return S


for i, phase in enumerate(['G1/S', 'S', 'G2', 'G2/M']):
    group = T[T['phase'] == phase]
    phase_genes = [x.upper() for x in group['Gene symbol'].unique()]

    # filter normalized expression
    pdf = ddf[ddf['geneName'].isin(phase_genes) & (ddf['control'] == 'timecourse')]

    mat = pd.pivot_table(pdf, 
                         index=['geneName'],
                         columns='hours',
                         values='value')

    scaler = StandardScaler()
    mat_norm = scaler.fit_transform(mat.T).T
    mat_norm = pd.DataFrame(mat_norm, 
                            index=mat.index, 
                            columns=mat.columns)

    sns.heatmap(mat_norm, 
                ax=axs[i, 0],
                cbar=False,
                cmap='RdYlGn')

    axs[i, 0].set_ylabel("")
    axs[i, 0].set_title(phase)
    axs[i, 0].set_xlabel("Time (hrs)")

    # smooth the signal
    smooth = smooth_rows(mat_norm, n_points=50, power=3)
    smooth = pd.melt(smooth)
    smooth['hours'] = smooth['variable'] * 8


    sns.lineplot(data=smooth,
                 x='hours',
                 y='value',
                 ax=axs[i, 1])

    axs[i, 1].set_ylabel("Mean Expression")
    axs[i, 1].set_yticks([])
    axs[i, 1].set_title(phase)
    axs[i, 1].set_xlabel("Time (hrs)")


sns.despine()
plt.tight_layout()
```

```python
cc_genes = s + g2
print(f"{len(cc_genes)=}")

cc = ddf[ddf['geneName'].isin(cc_genes)].reset_index()
non_cc = ddf[~ddf['geneName'].isin(cc_genes)].reset_index()

print(f"{cc.shape=} {non_cc.shape=}")

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.figsize'] = 5, 3

def meta_smoother(df):
    """a function to smooth a bunch of signals from
    tall data """
    pdf = df.copy()

    pdf = pdf[(pdf['control'] == 'timecourse')]
    mat = pd.pivot_table(pdf, 
                         index=['geneName'],
                         columns='hours',
                         values='value')

    # scaler = StandardScaler()
    # mat_norm = scaler.fit_transform(mat.T).T
    # mat_norm = pd.DataFrame(mat, 
    #                         index=mat.index, 
    #                         columns=mat.columns)

    smooth = smooth_rows(mat, n_points=50, power=3, standardize=False)
    smooth = pd.melt(smooth)
    smooth['hours'] = smooth['variable'] * 8

    return smooth


cc = meta_smoother(cc)
non_cc = meta_smoother(non_cc)


sns.lineplot(data=cc, 
             x='hours',
             y='value',
             label='Cell Cycle Genes')

sns.lineplot(data=non_cc, 
             x='hours',
             y='value',
             label='Non-Cell Cycle Genes')

plt.ylabel('Normalized Expression')
plt.xlabel('Time (hrs)')

sns.despine()
```
