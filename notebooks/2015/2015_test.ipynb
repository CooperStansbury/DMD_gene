```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
fpath = "../../data/raw_data/2015_tpm.csv"
df = pd.read_csv(fpath)
df['gene_mean'] = df.groupby('gene_name')['tpm'].transform('mean')
df['gene_std'] = df.groupby('gene_name')['tpm'].transform('std')

df['gene_norm'] = (df['tpm'] - df['gene_mean']) / df['gene_std']

df.head()
```

```python
query = ["CD38", 
         "FGFBP1", 
         "FGFBP2", 
         "PROM1"]

pdf = df[df['gene_name'].isin(query)]
pdf

sns.lineplot(data=pdf, 
             x='hours',
             y='gene_norm',
             hue='gene_name')
```
